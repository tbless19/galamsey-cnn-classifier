```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import seaborn as sns

# Spark imports
from pyspark.sql import SparkSession
from pyspark.ml.image import ImageSchema
from pyspark.sql.functions import col, udf
from pyspark.sql.types import ArrayType, FloatType, StringType

# TensorFlow/Keras imports
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
```

```python
class Config:
    """Configuration parameters for the project"""
    
    # Paths
    BASE_DIR = "galamsey"  # Base directory with yes/no subdirectories
    YES_DIR = os.path.join(BASE_DIR,"data/yes")
    NO_DIR = os.path.join(BASE_DIR,"data/no")
    OUTPUT_DIR = "output"
    PATCHES_DIR = os.path.join(OUTPUT_DIR, "patches")
    MODELS_DIR = os.path.join(OUTPUT_DIR, "models")
    RESULTS_DIR = os.path.join(OUTPUT_DIR, "results")
    
    # Image parameters
    PATCH_SIZE = 128
    CHANNELS = 3
    INPUT_SHAPE = (PATCH_SIZE, PATCH_SIZE, CHANNELS)
    
    # Data split ratios
    TRAIN_RATIO = 0.70
    VAL_RATIO = 0.15
    TEST_RATIO = 0.15
    
    # Training parameters
    BATCH_SIZE = 32
    EPOCHS = 10
    LEARNING_RATE = 0.001
    PATIENCE = 3
    
    # Augmentation parameters
    AUGMENTATION_FACTOR = 3  # Generate 3 augmented versions per patch
    
    # Model parameters
    DROPOUT_RATE = 0.5
    L2_REG = 0.01
    

    def create_directories():
        """Create necessary output directories"""
        for directory in [Config.OUTPUT_DIR, Config.PATCHES_DIR, 
                         Config.MODELS_DIR, Config.RESULTS_DIR]:


print("=" * 80)
print("SAT-5165 Final Project: Galamsey Image Classification")
print("=" * 80)

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# Create directories
Config.create_directories()

# Initialize Spark
print("\n[1/9] Initializing Spark Session...")
spark = SparkSession.builder \
    .appName("GalamseyClassification") \
    .config("spark.driver.memory", "4g") \
    .config("spark.executor.memory", "4g") \
    .getOrCreate()

print(f"Spark version: {spark.version}")
```

```python
# ============================================================================
# 2. DATA PREPROCESSING: PATCH EXTRACTION
# ============================================================================

class ImagePatchExtractor:
    """Extract patches from large images while preserving train/val/test split integrity"""
    
    def __init__(self, patch_size=128, overlap=0):
        self.patch_size = patch_size
        self.overlap = overlap
        self.stride = patch_size - overlap
    
    def extract_patches_from_image(self, image_path, label):
        """Extract non-overlapping patches from a single image"""
        patches = []
        img = Image.open(image_path).convert('RGB')
        img_array = np.array(img)
        
        h, w = img_array.shape[:2]
        
        # Extract patches in a grid
        for i in range(0, h - self.patch_size + 1, self.stride):
            for j in range(0, w - self.patch_size + 1, self.stride):
                patch = img_array[i:i+self.patch_size, j:j+self.patch_size]
                if patch.shape[:2] == (self.patch_size, self.patch_size):
                    patches.append({
                        'patch': patch,
                        'label': label,
                        'source_image': os.path.basename(image_path),
                        'position': (i, j)
                    })
        
        return patches
    
    def extract_patches_from_directory(self, directory, label):
        """Extract patches from all images in a directory"""
        all_patches = []
        image_files = [f for f in os.listdir(directory) 
                      if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        
        print(f"Processing {len(image_files)} images from {directory}...")
        
        for idx, img_file in enumerate(image_files):
            img_path = os.path.join(directory, img_file)
            patches = self.extract_patches_from_image(img_path, label)
            all_patches.extend(patches)
            
            if (idx + 1) % 10 == 0:
                print(f"  Processed {idx + 1}/{len(image_files)} images, "
                      f"extracted {len(all_patches)} patches so far")
        
        return all_patches
    
    def split_by_source_image(self, patches, train_ratio, val_ratio, test_ratio):
        """Split patches ensuring patches from same source image stay together"""
        # Group patches by source image
        image_groups = {}
        for patch in patches:
            source = patch['source_image']
            if source not in image_groups:
                image_groups[source] = []
            image_groups[source].append(patch)
        
        # Shuffle source images
        source_images = list(image_groups.keys())
        np.random.shuffle(source_images)
        
        # Calculate split indices
        n_images = len(source_images)
        train_end = int(n_images * train_ratio)
        val_end = train_end + int(n_images * val_ratio)
        
        # Split by source image
        train_sources = source_images[:train_end]
        val_sources = source_images[train_end:val_end]
        test_sources = source_images[val_end:]
        
        # Collect patches for each split
        train_patches = [p for src in train_sources for p in image_groups[src]]
        val_patches = [p for src in val_sources for p in image_groups[src]]
        test_patches = [p for src in test_sources for p in image_groups[src]]
        
        return train_patches, val_patches, test_patches

# Extract patches
print("\n[2/9] Extracting patches from images...")
extractor = ImagePatchExtractor(patch_size=Config.PATCH_SIZE)

galamsey_patches = extractor.extract_patches_from_directory(
    Config.YES_DIR, label=1
)
non_galamsey_patches = extractor.extract_patches_from_directory(
    Config.NO_DIR, label=0
)

print(f"\nExtracted {len(galamsey_patches)} galamsey patches")
print(f"Extracted {len(non_galamsey_patches)} non-galamsey patches")

# Split data by source image
print("\n[3/9] Splitting data (by source image)...")
all_patches = galamsey_patches + non_galamsey_patches
train_patches, val_patches, test_patches = extractor.split_by_source_image(
    all_patches, Config.TRAIN_RATIO, Config.VAL_RATIO, Config.TEST_RATIO
)

print(f"Train: {len(train_patches)} patches")
print(f"Val: {len(val_patches)} patches")
print(f"Test: {len(test_patches)} patches")
```

```python
class DataAugmentor:
    """Apply data augmentation to increase dataset size"""
    
    def __init__(self):
        self.augmentor = ImageDataGenerator(
            rotation_range=20,
            width_shift_range=0.2,
            height_shift_range=0.2,
            horizontal_flip=True,
            vertical_flip=True,
            zoom_range=0.2,
            shear_range=0.2,
            brightness_range=[0.8, 1.2],
            fill_mode='reflect'
        )
    
    def augment_patch(self, patch_array, n_augmentations=3):
        """Generate augmented versions of a single patch"""
        augmented = [patch_array]  # Include original
        
        # Reshape for ImageDataGenerator
        patch_batch = np.expand_dims(patch_array, axis=0)
        
        # Generate augmented versions
        aug_iter = self.augmentor.flow(patch_batch, batch_size=1)
        for _ in range(n_augmentations):
            augmented.append(next(aug_iter)[0].astype(np.uint8))
        
        return augmented
    
    def augment_patches(self, patches, n_augmentations=3):
        """Augment all patches in the list"""
        augmented_patches = []
        
        print(f"Augmenting {len(patches)} patches...")
        for idx, patch_dict in enumerate(patches):
            augmented_versions = self.augment_patch(
                patch_dict['patch'], n_augmentations
            )
            
            for aug_idx, aug_patch in enumerate(augmented_versions):
                augmented_patches.append({
                    'patch': aug_patch,
                    'label': patch_dict['label'],
                    'source_image': patch_dict['source_image'],
                    'augmentation_id': aug_idx
                })
            
            if (idx + 1) % 100 == 0:
                print(f"  Augmented {idx + 1}/{len(patches)} patches")
        
        return augmented_patches
            # Apply data augmentation (only to training set)
print("\n[4/9] Applying data augmentation...")
augmentor = DataAugmentor()
train_patches_augmented = augmentor.augment_patches(
    train_patches, n_augmentations=Config.AUGMENTATION_FACTOR
)

print(f"Training set after augmentation: {len(train_patches_augmented)} patches")
```

```python
# ============================================================================
# 4. SPARK DATA LOADING
# ============================================================================

class SparkDataLoader:
    """Load and preprocess images using Apache Spark"""
    
    def __init__(self, spark_session):
        self.spark = spark_session
    
    def save_patches_for_spark(self, patches, split_name):
        """Save patches as individual image files for Spark loading"""
        split_dir = os.path.join(Config.PATCHES_DIR, split_name)
        
        for label_name in ['galamsey', 'non_galamsey']:
            label_dir = os.path.join(split_dir, label_name)
            os.makedirs(label_dir, exist_ok=True)
        
        print(f"Saving {len(patches)} patches for {split_name} split...")
        
        for idx, patch_dict in enumerate(patches):
            label_name = 'galamsey' if patch_dict['label'] == 1 else 'non_galamsey'
            patch_path = os.path.join(
                split_dir, label_name, f"patch_{idx:05d}.png"
            )
            Image.fromarray(patch_dict['patch']).save(patch_path)
        
        return split_dir
    
    def load_images_with_spark(self, directory):
        """Load images using Spark's ImageSchema"""
        print(f"Loading images from {directory} using Spark...")
        
        df = self.spark.read.format("image").load(directory)
        
        # Extract label from path
        def extract_label(path):
            return 1 if 'galamsey' in path else 0
        
        extract_label_udf = udf(extract_label, FloatType())
        df = df.withColumn('label', extract_label_udf(col('image.origin')))
        
        return df
    
    def prepare_data_for_training(self, df):
        """Convert Spark DataFrame to numpy arrays for training"""
        data = df.collect()
        
        images = []
        labels = []
        
        for row in data:
            # Extract image data
            img_data = row['image']
            img_array = np.array(img_data.data).reshape(
                img_data.height, img_data.width, img_data.nChannels
            )
            images.append(img_array)
            labels.append(row['label'])
        
        return np.array(images), np.array(labels)
# Prepare data for training (convert to numpy arrays)
print("\n[5/9] Preparing data arrays...")

def patches_to_arrays(patches):
    X = np.array([p['patch'] for p in patches])
    y = np.array([p['label'] for p in patches])
    return X, y
    
X_train, y_train = patches_to_arrays(train_patches_augmented)
X_val, y_val = patches_to_arrays(val_patches)
X_test, y_test = patches_to_arrays(test_patches)

print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_val shape: {X_val.shape}, y_val shape: {y_val.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
```


## 5. CNN MODEL ARCHITECTURE

```python
class GalamseyClassifier:
    """CNN model for galamsey image classification"""
    
    def __init__(self, input_shape=(128, 128, 3), learning_rate=0.001):
        self.input_shape = input_shape
        self.learning_rate = learning_rate
        self.model = None
        self.history = None
    
    def build_model(self):
        """Build the CNN architecture as specified in the proposal"""
        
        model = models.Sequential([
            # Block 1
            layers.Conv2D(32, (3, 3), activation='relu', 
                         padding='same', input_shape=self.input_shape),
            layers.BatchNormalization(),
            layers.MaxPooling2D((2, 2)),
            
            # Block 2
            layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
            layers.BatchNormalization(),
            layers.MaxPooling2D((2, 2)),
            layers.Dropout(0.25),
            
            # Block 3
            layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
            layers.BatchNormalization(),
            layers.MaxPooling2D((2, 2)),
            layers.Dropout(0.25),
            
            # Dense layers
            layers.Flatten(),
            layers.Dense(128, activation='relu',
                        kernel_regularizer=keras.regularizers.l2(Config.L2_REG)),
            layers.Dropout(Config.DROPOUT_RATE),
            
            # Output layer
            layers.Dense(1, activation='sigmoid')
        ])
        
        # Compile model
        model.compile(
            optimizer=optimizers.Adam(learning_rate=self.learning_rate),
            loss='binary_crossentropy',
            metrics=['accuracy', 
                    keras.metrics.Precision(name='precision'),
                    keras.metrics.Recall(name='recall')]
        )
        
        self.model = model
        return model
        
    
    def train(self, X_train, y_train, X_val, y_val):
        """Train the model with callbacks"""
        
        # Callbacks
        early_stopping = EarlyStopping(
            monitor='val_loss',
            patience=Config.PATIENCE,
            restore_best_weights=True,
            verbose=1
        )
        
        model_checkpoint = ModelCheckpoint(
            os.path.join(Config.MODELS_DIR, 'best_model.h5'),
            monitor='val_accuracy',
            save_best_only=True,
            verbose=1
        )
        
        reduce_lr = ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=2,
            min_lr=1e-7,
            verbose=1
        )
        
        # Normalize images
        X_train_norm = X_train.astype('float32') / 255.0
        X_val_norm = X_val.astype('float32') / 255.0
        
        # Train
        self.history = self.model.fit(
            X_train_norm, y_train,
            validation_data=(X_val_norm, y_val),
            batch_size=Config.BATCH_SIZE,
            epochs=Config.EPOCHS,
            callbacks=[early_stopping, model_checkpoint, reduce_lr],
            verbose=1
        )
        
        return self.history    
    
def evaluate(self, X_test, y_test):
        """Evaluate model on test set"""
        X_test_norm = X_test.astype('float32') / 255.0
        
        results = self.model.evaluate(X_test_norm, y_test, verbose=0)
        
        # Get predictions
        y_pred_proba = self.model.predict(X_test_norm)
        y_pred = (y_pred_proba > 0.5).astype(int).flatten()
        
        return {
            'test_loss': results[0],
            'test_accuracy': results[1],
            'test_precision': results[2],
            'test_recall': results[3],
            'predictions': y_pred,
            'probabilities': y_pred_proba.flatten()
        }
# Build and train model
print("\n[6/9] Building CNN model...")
classifier = GalamseyClassifier(
    input_shape=Config.INPUT_SHAPE,
    learning_rate=Config.LEARNING_RATE
)
classifier.build_model()
classifier.model.summary()
```

```python
print("\n[7/9] Training model...")
history = classifier.train(X_train, y_train, X_val, y_val)
```

```python
def evaluate(X_test, y_test):
        """Evaluate model on test set"""
        X_test_norm = X_test.astype('float32') / 255.0
        
        results = classifier.model.evaluate(X_test_norm, y_test, verbose=0)
        
        # Get predictions
        y_pred_proba = classifier.model.predict(X_test_norm)
        y_pred = (y_pred_proba > 0.5).astype(int).flatten()
        
        return {
            'test_loss': results[0],
            'test_accuracy': results[1],
            'test_precision': results[2],
            'test_recall': results[3],
            'predictions': y_pred,
            'probabilities': y_pred_proba.flatten()
        }
# Evaluate on test set
print("\n[8/9] Evaluating on test set...")
test_results = evaluate(X_test, y_test)

print(f"\nTest Results:")
print(f"  Loss: {test_results['test_loss']:.4f}")
print(f"  Accuracy: {test_results['test_accuracy']:.4f}")
print(f"  Precision: {test_results['test_precision']:.4f}")
print(f"  Recall: {test_results['test_recall']:.4f}")

# Calculate F1 score
precision = test_results['test_precision']
recall = test_results['test_recall']
f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
print(f"  F1 Score: {f1_score:.4f}")
```

```python
# ============================================================================
# 6. VISUALIZATION AND EVALUATION
# ============================================================================

class ResultsVisualizer:
    """Visualize training results and model performance"""
    
    @staticmethod
    def plot_training_history(history, save_path):
        """Plot training and validation metrics"""
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        
        # Accuracy
        axes[0, 0].plot(history.history['accuracy'], label='Train')
        axes[0, 0].plot(history.history['val_accuracy'], label='Validation')
        axes[0, 0].set_title('Model Accuracy')
        axes[0, 0].set_xlabel('Epoch')
        axes[0, 0].set_ylabel('Accuracy')
        axes[0, 0].legend()
        axes[0, 0].grid(True)
        
        # Loss
        axes[0, 1].plot(history.history['loss'], label='Train')
        axes[0, 1].plot(history.history['val_loss'], label='Validation')
        axes[0, 1].set_title('Model Loss')
        axes[0, 1].set_xlabel('Epoch')
        axes[0, 1].set_ylabel('Loss')
        axes[0, 1].legend()
        axes[0, 1].grid(True)
        
        # Precision
        axes[1, 0].plot(history.history['precision'], label='Train')
        axes[1, 0].plot(history.history['val_precision'], label='Validation')
        axes[1, 0].set_title('Model Precision')
        axes[1, 0].set_xlabel('Epoch')
        axes[1, 0].set_ylabel('Precision')
        axes[1, 0].legend()
        axes[1, 0].grid(True)
        
        # Recall
        axes[1, 1].plot(history.history['recall'], label='Train')
        axes[1, 1].plot(history.history['val_recall'], label='Validation')
        axes[1, 1].set_title('Model Recall')
        axes[1, 1].set_xlabel('Epoch')
        axes[1, 1].set_ylabel('Recall')
        axes[1, 1].legend()
        axes[1, 1].grid(True)
        
        plt.tight_layout()
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Training history plot saved to {save_path}")
        plt.show


    def plot_confusion_matrix(self, y_true, y_pred, save_path):
        """Plot confusion matrix"""
        cm = confusion_matrix(y_true, y_pred)
        
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                   xticklabels=['Non-Galamsey', 'Galamsey'],
                   yticklabels=['Non-Galamsey', 'Galamsey'])
        plt.title('Confusion Matrix')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.tight_layout()
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"Confusion matrix saved to {save_path}")
        plt.show()


    def plot_roc_curve(self,y_true, y_pred_proba, save_path):
        """Plot ROC curve and calculate AUC"""
        fpr, tpr, thresholds = roc_curve(y_true, y_pred_proba)
        roc_auc = auc(fpr, tpr)
        
        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, color='darkorange', lw=2,
                label=f'ROC curve (AUC = {roc_auc:.3f})')
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--',
                label='Random Classifier')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Receiver Operating Characteristic (ROC) Curve')
        plt.legend(loc="lower right")
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"ROC curve saved to {save_path}")
        plt.show()
        
        return roc_auc

    def generate_classification_report(self,y_true, y_pred, save_path):
        """Generate and save classification report"""
        report = classification_report(
            y_true, y_pred,
            target_names=['Non-Galamsey', 'Galamsey'],
            digits=4
        )
        
        with open(save_path, 'w') as f:
            f.write("Classification Report\n")
            f.write("=" * 50 + "\n\n")
            f.write(report)
        
        print(f"Classification report saved to {save_path}")
        print("\n" + report)

        
```
